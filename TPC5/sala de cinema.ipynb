```python
#Modelo 
# Cinema = [Sala]
# Sala = [nlugares, Vendidos, filme, sala]
# nlugares = Int
# filme = String 
# Vendidos = [Int]

def inserirFilmes (cinema1, sala):
    cinema1.append (sala)
    return cinema1 

def listar (cinema1):
    print (".........Filmes Disponíveis.........")
    for sala in cinema1:
        print ( "--" , sala[2], "--")
    return cinema1

def disponivel (cinema1, filme, lugar):
    condição = False
    for sala in cinema1:
        if sala[2] == filme:
            if lugar not in sala[1]:
                condição = True
    return condição

def vendeBilhete (cinema1, filme, lugar):
    if disponivel (cinema1, filme, lugar):
        for sala in cinema1:
            if sala[2] == filme:
                sala[1].append (lugar)
        return cinema1
    
def listarDisponibilidades (cinema1):
    print ("..Exibindo..")
    for sala in cinema1:
      lugaresdisponiveis = sala [0] - len (sala[1])
      print (sala[3], "Nome:", sala[2], "--Lugares Disponíveis--", lugaresdisponiveis )
    return cinema1 

def devolverBilhetes (cinema1, filme, lugar):
    for sala in cinema1:
        if sala[2] == filme:
            sala[1].remove (lugar)
        return cinema1
    
def listarSala (cinema1, filme):
    for sala in cinema1:
        if sala[2] == filme:
            print ("Filme:", sala[2], "-- Número de lugares:", sala[0], "-- Lugares ocupados:", sala[1])
        return 
 
def criarSala (cinema1, filme, lugar):
    sala = (lugar, [], filme)
    cinema1.append (sala)
    return cinema1

def removerSala (cinema1, filme, lugar):
    for sala in cinema1:
        if sala[2] == filme:
            if sala[1] == []:
                cinema1.remove (sala)
            return cinema1
        


  


def menu():
    print(""" 
    (1) Criar sala
    (2) Remover sala
    (3) Listar sala
    (4) Cartaz em exibição
    (5) Listar as disponibilidades do Cinema1
    (6) Consulta Lugares
    (7) Venda de bilhetes
    (8) Devolução de bilhetes
    (0) Sair
    """)
```
