```python
# Derivada do Polinómio

def calculader (polinomio):
    derivada = []
    for coeficiente, grau in polinomio:
        coeficiente = coeficiente * grau
        grau = grau -1
        derivada.append ((coeficiente , grau))
        if coeficiente == 0:
            derivada.remove ((coeficiente ,  grau))
    return derivada

calculader (p1)
```

```python
# Soma de dois polinómios

def somapol (polinomio1,polinomio2):
    soma = []
    i = 0
    j = 0
    while i < len(polinomio1) and j < len(polinomio2):
        coeficiente1, grau1 = polinomio1[i]
        coeficiente2, grau2 = polinomio2[j]
        if grau1 == grau2:
            coeficiente = coeficiente1 + coeficiente2
            grau = grau1 = grau2
            soma.append ((coeficiente, grau))
        elif grau1 > grau2:
            soma.append ((coeficiente1, grau1))
            i = i + 1
        elif grau1 < grau2:
            soma.append ((coeficiente2, grau2))
            j = j + 1

    return soma


```

```python
# A plicação para manipular polinómio
# Criar um polinómio

def criaPolin ():
    grau = int(input("Introduza o grau do polinómio que deseja fazer"))
    polinomio = []
    while grau >= 0:
        coeficiente = int(input("Introduza o o coeficiente do grau" +  str(grau)))
        if coeficiente != 0:
            termo = (coeficiente, grau)
            polinomio.append (termo)
            grau = grau - 1
    return polinomio
```

```python
# Ler uma lista de polinómios de um ficheiro

def lerPol (ficheiro):
    lista = []
    file = open(ficheiro , "r")
    for line in file:
        termos = line.split (":")
        polinomio = []
        for termo in termos:
            x = termo.split (";")
            if len(x) > 1:
                coeficiente = x[0]
                grau = x[1]
                polinomio.append ((coeficiente , grau))
        lista.append (polinomio)
    return lista
```

```python
# Listar polinómios

def listarPol (lista):
    polinomio = {}
    i = 1
    while i < len(lista):
        polinomio[i] = lista [i - 1]
        i = i +1
    print ("Númeor de ordem ---- Polinómio")
    print ("......................")
    for chave in polinomio:
        print (chave, "|", polinomio[chave])
    return 
```

```python
# Calcular o valor de um polinómio num ponto

def pot (a,b):
    res = 1
    while b > 0:
        res = res * a
        b = b + 1
    return res


def polPonto (p,x):
    res = 0
    for termo in p:
        coeficiente, grau = termo
        res = res + coeficiente * pot (x, grau)
    return res 
```

```python
# Listar polinómios com grau

def listarPolGrau (lista):
    polinomio = {}
    i = 1
    while i < len(lista):
        polinomio[i] = [lista[i-1], lista[i-1][0][1]]
        i = i + 1
    print ("Número de ordem --- Polinómio --- Grau")
    print (".......................................")
    for chave in polinomio:
        pol , grau = polinomio[chave]
        print (chave, ";" , pol, ";" , grau)
    return
```

```python
# Maior grau

def maiorGrau (lista):
    maior = 0
    polinomio = {}
    i = 1
    while i < len(lista):
        polinomio[i] = [lista[i-1], lista[i-1][0][1]]
        i = i + 1
        for chave , valor in polinomio.items():
            grau = valor [1]
            if grau > maior:
                maior = grau
                n = chave
    return (n , maior)
```

```python
# Gravar polinómios num ficheiro

def guardarPol (polinomio, fnome):
    ficheiro = open (fnome, "w")
    for p in polinomio:
        for termo in p:
            coeficiente , grau = termo
            ficheiro.write (str(coeficiente) + ";" + str(grau) + "|")
        ficheiro.write ("\n")
    ficheiro.close ()

guardarPol (polinomio, "polinomio aquecimento.txt")
```
