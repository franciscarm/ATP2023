```python


def distds (ficheiro):
    fem_doente = 0
    fem_ndoente = 0 
    masc_doente = 0
    masc_ndoente = 0
    fich = open (ficheiro)
    for linhas in fich:
        elemento = linhas.split (";")
        genero = elemento [0]
        doençacardíaca = elemento [3]
        if genero == "Female":
            if doençacardíaca == "0":
                fem_ndoente = fem_ndoente + 1
            if doençacardíaca != "1":
                fem_doente = fem_doente + 1
            if doençacardíaca == "0":
                masc_ndoente = masc_ndoente + 1
            if doençacardíaca == "1":
                masc_doente = masc_doente + 1
    total_female = fem_doente + fem_ndoente
    total_male = masc_doente + masc_ndoente
    distfem = float ((fem_doente/total_female)*100)
    distmasc = float ((masc_doente/total_male)*100)
    print ("A percentagem de mulheres doentes é:" + str (distfem) + "%")
    print ("A percentagem de homens doentes é:" + str (distmasc) + "%")
    fich.close ()
    return 

distds ("diabetes_prediction_dataset.csv")

   
```
